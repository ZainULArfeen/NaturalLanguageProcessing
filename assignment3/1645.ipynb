```python
import selenium
from selenium import webdriver
from bs4 import BeautifulSoup
import requests
import time
```

```python
driver = webdriver.Chrome()

#list for all products URL save
list1=[]

URL = 'https://www.daraz.pk/mens-t-shirts/'
for i in range(1,6): 
    URL = 'https://www.daraz.pk/mens-t-shirts/?page='+str(i) 
    driver.get(URL)  
    print(URL) 
    src=driver.page_source 
    soup=BeautifulSoup(src,'html.parser')
    time.sleep(5)
    urll=soup.findAll('div',{'class':'c16H9d'})
    #print(urll)
    for i in urll:
        i=i.find('a') 
        print(i['href']) 
        list1.append(i['href'])
        


driver.quit()
```

```python
urlList = list1
scrappedData = []
product = 0

darazData = open('tshirtdata.csv', 'a', encoding='utf-8')

darazData.write("product,productname,productprice,productreview\n")

driver = webdriver.Chrome()

for link in urlList:
    product += 1
    print("product : " + str(product) )
    
    link = link.replace('\n', '')
    URL = 'https:' + str(link)
    driver.get(URL)
    print(URL)
    src = driver.page_source
    
    soup = BeautifulSoup(src, 'html.parser')
    time.sleep(2)
    
    productname = ''
    
    if soup.find('span', {'class':'pdp-mod-product-badge-title'}):
        productname = soup.find('span', {'class':'pdp-mod-product-badge-title'})
        productname = productname.get_text()
        print(productname)
        scrappedData.append(productname)
        
        
        productprice = ''
    
    if soup.find('div', {'class':'pdp-product-price'}):
        productprice = soup.find('span', {'class':'pdp-price pdp-price_type_normal pdp-price_color_orange pdp-price_size_xl'})
        productprice = productprice.get_text()
        print("\nprice : " + str(productprice))
        scrappedData.append(productprice)
        
        
    
    if soup.find('div', {'class':'mod-reviews'}):
        reviews = soup.findAll('div', {'class':'content'})
        for review in reviews:
            review = review.get_text()
            review = review.replace('.', '')
            review = review.replace('\n', '')
            review = review.replace('\t', '')
            review = review.replace('\n\n', '')
            review = review.replace(',', '')
            review = review.replace('\r', '')
            review = review.strip()
            print("\nreview : " + str(review))
            scrappedData.append(review)
            
            
            
            
            darazData.write("product" + str(product)+ ',' +productname + ',' + productprice + ',' + str(review))
            darazData.write('\n')

    
driver.quit()
```

```python
# Removing Punctuations
lll = [''.join(c for c in s if c not in string.punctuation) for s in lll]
lll
```

```python
#removing Emojis
import emoji

def extract_emojis(s):
    return ''.join(c for c in s if c not in emoji.UNICODE_EMOJI)

no_Emo = []

for sent in lll:
    sent = extract_emojis(sent)
    no_Emo.append(sent)
```

```python
#positive lexicon

f = open('positiveword.txt', 'r')
pos_lex = list(f)
pos_lex = [s.strip('\n') for s in pos_lex]
pos_lex
```

```python
# negative lexicon

f = open('negativeword.txt', 'r')
neg_lex = list(f)
neg_lex = [s.strip('\n') for s in neg_lex]
neg_lex
```

```python
sentiments = []

for sent in no_Emo:
        for word in sent.split():
            positive = 0
            negative = 0
            for i in pos_lex:
                if word == i:
                    positive += 1
            for j in neg_lex:
                if j == word:
                    negative += 1
                    
        if positive > negative:
            sentiments.append("positive")
        if negative > positive:
            sentiments.append("negative")
        if positive == negative:
            sentiments.append("neutral")
            
```

```python
neutral_sent = sentiments.count('neutral')
pos_sent = sentiments.count('positive')
neg_sent = sentiments.count('negative')

print("neutral sentiments : " + str(neutral_sent) + "\npositive sentiments : " + str(pos_sent) + 
         "\nnegative sentiments : " + str(neg_sent) )
```
