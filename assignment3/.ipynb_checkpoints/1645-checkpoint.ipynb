```python
import selenium
from selenium import webdriver
from bs4 import BeautifulSoup
import requests
import time
```

```python
driver = webdriver.Chrome()

#list for all products URL save
list1=[]


#create CSV for Saving Data of mobile phones with sentiment polarity
#data=open('DarazDataComplete.csv','a',encoding="utf-8")

#writing header for the FILE
#data.write("ProductNumber,BrandName,BrandTitle,URL,AverageRating,Review\n")


URL = 'https://www.daraz.pk/mens-t-shirts/'
for i in range(1,6): 
    URL = 'https://www.daraz.pk/mens-t-shirts/?page='+str(i) 
    driver.get(URL)  
    print(URL) 
    src=driver.page_source 
    soup=BeautifulSoup(src,'html.parser')
    time.sleep(5)
    urll=soup.findAll('div',{'class':'c16H9d'})
    #print(urll)
    for i in urll:
        i=i.find('a') 
        print(i['href']) 
        list1.append(i['href'])
        


driver.quit()
```

```python
urlList = list1
scrappedData = []
product = 0

darazData = open('darazData.csv', 'a', encoding='utf-8')

darazData.write("Product,Product Name, Product Price, Product Review \n")

driver = webdriver.Chrome()

for link in urlList:
    product += 1
    print("product : " + str(product) )
    
    link = link.replace('\n', '')
    URL = 'https:' + str(link)
    driver.get(URL)
    print(URL)
    src = driver.page_source
    
    soup = BeautifulSoup(src, 'html.parser')
    time.sleep(2)
    
    productname = ''
    
    if soup.find('span', {'class':'pdp-mod-product-badge-title'}):
        productname = soup.find('span', {'class':'pdp-mod-product-badge-title'})
        productname = productname.get_text()
        print(productname)
        scrappedData.append(productname)
        
        
        productprice = ''
    
    if soup.find('div', {'class':'pdp-product-price'}):
        productprice = soup.find('div', {'class':'pdp-product-price'})
        productprice = productprice.get_text()
        print("\nprice : " + str(productprice))
        scrappedData.append(productprice)
        
        
    
    if soup.find('div', {'class':'mod-reviews'}):
        reviews = soup.findAll('div', {'class':'content'})
        for review in reviews:
            review = review.get_text()
            review = review.replace('.', '')
            review = review.replace('\n', '')
            review = review.replace('\t', '')
            review = review.replace('\n\n', '')
            review = review.replace(',', '')
            review = review.replace('\r', '')
            review = review.strip()
            print("\nreview : " + str(review))
            scrappedData.append(review)
            
            
            darazData.write("product" + str(product)+ ',' +productname + ',' + productprice + ',' + str(review))
            darazData.write('\n')

    
driver.quit()
```
